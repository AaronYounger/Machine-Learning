```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model

%matplotlib inline
peterrabit = pd.read_csv("Houseprices.csv")
d = pd.read_csv("areas.csv")


plt.xlabel('area')
plt.ylabel('price')
plt.scatter(peterrabit.area, peterrabit.price, color = 'red', marker = '*')
plt.show()

reg = linear_model.LinearRegression()
reg.fit(peterrabit[['area']], peterrabit.price)

plt.xlabel('area')
plt.ylabel('price')
plt.scatter(peterrabit.area, peterrabit.price, color = 'red', marker = '*')
plt.plot(peterrabit.area, reg.predict(peterrabit[['area']]), color = 'blue')
plt.show()

p = reg.predict(d)
d['prices'] = p
d.to_csv('predictions.csv')

df1 = pd.read_csv("predictions.csv")

reg2 = linear_model.LinearRegression()
reg2.fit(df1[['area']], df1.prices)

plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df1.area, df1.prices, color = 'red', marker = "*")
plt.plot(df1.area, reg2.predict(df1[['area']]), color = 'blue')
plt.show()

y_true = peterrabit.price
y_pred = reg.predict(peterrabit[['area']])

r2 = r2_score(y_true, y_pred)
mae = mean_absolute_error(y_true, y_pred)
rmse = np.sqrt(mean_squared_error(y_true, y_pred))

print("In-sample R:", r2)
print("In-sample MAE:", mae)
print("In-sample RMSE:", rmse)



```
