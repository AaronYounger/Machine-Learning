```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
import warnings
warnings.filterwarnings("ignore")

np.random.seed(123)

df = pd.read_csv('77_Close_Houses2.csv')
df.head()
reg = linear_model.LinearRegression()

df = df.rename(columns={
    'BATHS': 'baths',
    'BEDS': 'beds',
    'SQUARE FEET': 'sqft',
    'LOT SIZE': 'lot_size',
    'YEAR BUILT': 'year_built',
    'PRICE': 'price'
})


df = df.dropna(subset=['baths','beds','sqft','lot_size','year_built','price'])
X = df[['baths','beds','sqft','lot_size','year_built']]
y = df['price']
reg.fit(X,y)
my_property = [[2,3, 1280, 7405.2, 1972]]
score = reg.score(X,y)

original_price = 295000
prediction = reg.predict(my_property)

pred_value = prediction[0]   # extract from array
difference = original_price - pred_value

print(f"Predicted House Price: ${pred_value:,.2f}")
print(f"House Price: ${original_price:,.2f}")
print(f"Overpriced by: ${difference:,.2f}")
print(f"Model Score: {score*100:,.2f}%")
```
