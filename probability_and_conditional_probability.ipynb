```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("dataset-01.csv")

count_total = len(df)

# Probability of Students who got an 'A' in OOP, 'A' in Java

X_a_OOP = (df["OOP"]=='A').sum()
P_A_OOP = (X_a_OOP/count_total)*100
print("Probability of A in OOP:", P_A_OOP,"%")

Y_a_Java = (df["Java"]=='A').sum()
P_A_Java = (Y_a_Java/count_total)*100
print("Probability of A in Java:", P_A_Java,"%")

# Find the probability of P(A|B)
# Probability of X = A in OOP and Y = B in Java

# Calculuate Numerator
countof_A_intersection_B = 0

for index, row in df.iterrows():
  if row['Java']== 'B' and row['OOP']== 'A':
    countof_A_intersection_B += 1
print("Count of A in OOP and B in Java:", countof_A_intersection_B)

# Probability of A intersection B

pA_intersection_B = countof_A_intersection_B/count_total
print("Probability of A intersection B:", pA_intersection_B)

# Calculate Denominator
Y_B_Java = (df["Java"]=='B').sum()
P_B_Java = Y_B_Java/count_total
print("Probability of B in Java:", P_B_Java)

# Find conditional Probability of A given B

condition_A_given_B = pA_intersection_B/P_B_Java
print("The Conditional Probability of A given B:", condition_A_given_B)

# Among the students who got a B in Java, there is a 50% chance they got an A in OOP

```
