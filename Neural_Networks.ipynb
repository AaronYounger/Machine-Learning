```python
import numpy as np
import pandas as pd

np.random.seed(0)

df = pd.read_csv("dataset.csv")
x = df[['a','b']]
# y = df['C']

class Layer_Dense:
  def __init__(self, n_inputs, n_neurons):
    self.weights = 0.1 * np.random.randn(n_inputs, n_neurons)
    self.biases = np.zeros((1, n_neurons))

  def forward(self, inputs):
    self.output = np.dot(inputs, self.weights) + self.biases
    return self.output

class Activation_ReLU:
    def forward(self, inputs):
      self.output = np.maximum(0, inputs)
      return self.output
# return instead of print so I can keep using this function within the code


layer1 = Layer_Dense(2, 5)
aa1 = Activation_ReLU()
layer2 = Layer_Dense(5, 3)
aa2 = Activation_ReLU()
layer3 = Layer_Dense(3, 1)

# Forward Pass

layer1 = layer1.forward(x)
aa1 = aa1.forward(layer1)

layer2 = layer2.forward(aa1)
aa2 = aa2.forward(layer2)

layer3 = layer3.forward(aa2)

print("Layer1 output:\n", layer1)
print("Layer2 output:\n", layer2)
print("Layer3 output:\n", layer3)

```
